```python
import pandas as pd
```

```python
all_data_df = pd.read_pickle('../dump/all_data')
```

```python
all_data_df
```

```python
all_df = all_data_df.drop(columns=['vote','metascore','keywords',\
                                   'link_d','link_w',\
                                   'book_title','rating_value','rating_count','review_count','title']).\
                    drop_duplicates(subset=['movie_title','director'])
all_df.rename(columns = {'certificate':'MPAA','star':'actor','year':'publish_year'},inplace=True)
```

```python
all_df = all_df.dropna().reset_index()
```

```python
all_df.shape
```

```python
# Create list of actor/link from the 500 movies
star_list = []
for star in all_df.actor:
    star_list += star
star_df = pd.DataFrame(star_list)

star_link_list = []
for star_link in all_df.link_s:
    star_link_list += star_link
star_link_df = pd.DataFrame(star_link_list)

# Combine in to one df
star_df['link_s'] = star_link_df
star_df.columns = ['star','link_s']
star_df = star_df.drop_duplicates(subset=['star'])
```
